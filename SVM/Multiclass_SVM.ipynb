```python
import pandas as pd  # frames
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv('../Preprocessing_pipeline/train_data.csv')
train_data.head()
```

```python
categories_platform_used = train_data["Platforms Used"].unique()
for i in range(len(categories_platform_used)):
    categories_platform_used[i] = categories_platform_used[i].split(",")

print(categories_platform_used)
print(categories_platform_used[0])
```

```python
social_media_platforms = []
for i in categories_platform_used:
    for j in i:
        if j not in social_media_platforms:
            social_media_platforms.append(j)

# Removing duplicate values because of text formatting
to_remove = [11, 12, 10, 13, 9, 14, 15]

social_media_platforms = [
    i for j, i in enumerate(social_media_platforms) if j not in to_remove
]

# remove prior blank spaces
for i in range(len(social_media_platforms)):
    social_media_platforms[i] = social_media_platforms[i].strip()

print("Number of social media platforms =", len(social_media_platforms))
print(social_media_platforms)
```
